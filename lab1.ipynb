```python
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
categorical_columns = df.select_dtypes(include='object').columns
```

```python
label_encode_columns = []
one_hot_encode_columns = []

for column in df.columns:
    unique_values = df[column].nunique()
    if unique_values == 2:
        label_encode_columns.append(column)
    elif unique_values > 2 and unique_values < 5:
        one_hot_encode_columns.append(column)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for column in label_encode_columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
df.head()
```

```python
one_hot_encode_columns
```

```python
df = pd.get_dummies(data=df, columns=one_hot_encode_columns)
```

```python
import xgboost as xgb
```

```python
object_columns = df.select_dtypes(include='object').columns
df['Total_Payment'] = df['Total_Payment'].replace(r'^\s*$', float('nan'), regex=True)
df['Total_Payment'] = pd.to_numeric(df['Total_Payment'], errors='coerce')
```

```python
X = df.drop('Leave_Next_Month',  axis=1)
y = df['Leave_Next_Month']
```

```python
dtrain = xgb.DMatrix(data=X, label=y)
```

```python
params = {"objective":"binary:logistic",'colsample_bytree': 0.4,'learning_rate': 0.1,
                'max_depth': 5, 'alpha': 10}
res = xgb.cv(
    params,
    dtrain,
    num_boost_round=10,
    nfold=5,
    metrics="auc",
    seed=0,
    callbacks=[
        xgb.callback.EvaluationMonitor(show_stdv=False),
        xgb.callback.EarlyStopping(3),
    ],
)
best_num_boost_rounds = res.shape[0]
```

```python
categorical_columns = df_test.select_dtypes(include='object').columns
label_encode_columns = []
one_hot_encode_columns = []

for column in df_test.columns:
    unique_values = df_test[column].nunique()
    if unique_values == 2:
        label_encode_columns.append(column)
    elif unique_values > 2 and unique_values < 5:
        one_hot_encode_columns.append(column)
label_encoder = LabelEncoder()
for column in label_encode_columns:
    df_test[column] = label_encoder.fit_transform(df_test[column])
df_test = pd.get_dummies(data=df_test, columns=one_hot_encode_columns)
df_test['Total_Payment'] = df_test['Total_Payment'].replace(r'^\s*$', float('nan'), regex=True)
df_test['Total_Payment'] = pd.to_numeric(df_test['Total_Payment'], errors='coerce')
```

```python

best_num_boost_rounds = res.shape[0]
X_test = df_test
dtest = xgb.DMatrix(X_test)

model = xgb.train(params, dtrain, num_boost_round=best_num_boost_rounds)

y_pred = model.predict(dtest)
```

```python
result = pd.DataFrame({'id': df_test['id'], 'Leave_Next_Month': y_pred})
result.to_csv('predictions.csv', index=False)
```
